```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
token_id=39
layer_id=38
num_token=40
seq=get_layer(layer_id)[token_id]
# seq[8:]=0
viewing=np.reshape(norm_l(seq)[:num_token],(1,-1))
df_cm = pd.DataFrame(
    viewing, index=[tokens[token_id]], columns=[f'{tok}:{idx}' for idx,tok in enumerate(tokens[:40])]
)
plt.figure(figsize=(10, 1))
sn.heatmap(df_cm, annot=False, cmap="Greens")
```
